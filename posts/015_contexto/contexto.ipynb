## Reflection
Well.. it took Chat **17** guesses to get to the right answer! Personally, I dont think this is very bad. If you have played the game before and done better, congrats! 

I find it very interesting that this specific model uses **SO MANY** emojis... to the point that it kinda gets on my nerves. 

I think the model does this to attempt to mimic younger audiences (specifically in the way that they text one another, but I'm not sure)

In addition, I found it pretty odd that it went from refined concepts to broader concepts. Usually when I play this game I start with very broad concepts and then narrow my scope from there (think emotion, object, food, etc). Chat went from **apple** to **fruit** and was surprised when it was further from the answer.
