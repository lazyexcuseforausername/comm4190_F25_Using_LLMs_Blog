## Exploring an LLM’s Myers–Briggs “Personality”

Welcome to this little experiment in machine “personality.” In this blog, I’m putting a large language model through the Myers–Briggs Type Indicator (MBTI) — the same personality test you’ve probably seen on TikTok, in icebreakers, or on way-too-detailed dating profiles. The core question: if we give an LLM the exact same questions humans get, what kind of “personality type” does it end up with, or will it refuse to even cooperate?

The Myers–Briggs test sorts people into 16 types based on four preference pairs:

- **Extraversion (E) vs. Introversion (I)** – where you focus your energy: outward (people, action) or inward (thoughts, reflection).  
- **Sensing (S) vs. Intuition (N)** – how you prefer to take in information: concrete details or big-picture patterns.  
- **Thinking (T) vs. Feeling (F)** – how you tend to make decisions: logic/consistency or values/impact on people.  
- **Judging (J) vs. Perceiving (P)** – how you like to organize your life: structured and planned or flexible and open-ended.

Put together, you get a four-letter “type” like **INTJ**, **ENFP**, or **ISFJ**, which many people treat as a loose personality map.

In this blog, I’ll walk through how an LLM responds to these questions, what type it gets, and how consistent those answers are if we repeat the test with a different model. Along the way, I’ll also talk about the limits of MBTI itself and why using it on an LLM is both a little cursed and also surprisingly revealing.

Answers to the far left are "agree" the middle is "neutral" and the far right is "disagree" for the purposes of the llm I will be assigning numerical values to the scale so I can accurately distinguish its answers.
